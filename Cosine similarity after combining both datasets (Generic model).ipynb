### Stemming Process

```python
import nltk #for stemming process
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#defining the stemming function
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
#applying stemming on the tags column
combined['tags'] = combined['tags'].apply(stem)  ##HAVE NOT REMOVED DUPLICATE WORDS FROM SET
```

# Cosine similarity measure (content similarity)

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(combined["tags"])

##Compute the Cosine Similarity for all courses to job title (last row) based on the count_matrix
cosine_sim = cosine_similarity(count_matrix) 
sim_df = pd.DataFrame(cosine_sim,index=combined.course_name,columns=combined.course_name)
sim_df.tail()
```

```python
#sort by most similar/ closest datapoint to target job title
recommended_courses= sim_df[target_job].sort_values(ascending=False)[:21]
recommended_courses=list(recommended_courses.index.values)
print('For the job', recommended_courses[0], 'the top 20 most relevant courses are:', recommended_courses[1:])
```
